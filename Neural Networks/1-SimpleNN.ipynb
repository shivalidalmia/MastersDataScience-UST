```python
# Importing libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# For having a single layer with single neuron, with one input shape
# Defining the architecture of the model
model = tf.keras.Sequential(
    [keras.layers.Dense(units=1, input_shape = [1])]
)
```

```python
# Compiling the model
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
# Dataset (y = 2*X + 2)
X = np.array([-2,-1,0,1,2,3,4], dtype =float)
y = np.array([-2,0,2,4,6,8,10], dtype=float)
```

```python
# Fitting the model on data no of epochs is a hyperparameter
model.fit(X, y, epochs=500)

# How to identify number of epochs - Uptill loss is going down. When loss stop going down that is the correct number of epochs when no more learning is to be done
```

```python
print(model.predict([9]))
```

```python
# Predicting cost of price with one person
# Dataset (y = 10K + 3k) where k is no of person
X = np.array([1,2,3,4,5,6,7], dtype =float)
y = np.array([13,16,19,22,25,28,31], dtype=float)
```

```python
# Fitting the model
model.fit(X, y, epochs = 500)
```
