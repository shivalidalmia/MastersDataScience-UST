```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
sentences = [
    'i like my house',
    'I, like my car',
    'You like my car!'
    
]

tokenizer = Tokenizer(num_words = 100) #parameter is used to utilize top 100 words that appear
tokenizer.fit_on_texts(sentences) #Look at all words and gives unique number to each word
word_index = tokenizer.word_index #Shows the  word index for each word
print(word_index)
```
